````python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
````

##  Visualize the data using countplot, histplot, boxplot, etc., on features: survived, class, sex, age, etc.,

````python
sns.countplot(x='survived', data= titanic)
plt.xlabel('Survival Status')
plt.ylabel('Number of People')
plt.xticks([0,1], ['Not Survived', 'Survived'])
#This will display the x-axis with labels "Not Survived" and "Survived" instead of the default 0 and 1 for the survival status.
plt.title('Survival Graph')
plt.show()
````

````python
sns.countplot(x= 'sex', hue= 'survived', data= titanic, palette='pastel')
plt.xlabel('Gender')
plt.ylabel('Number of People')
plt.legend(['Not Survived', 'Survived'])
plt.title('Survival Graph Based on Gender')
plt.show()


# x='sex': This specifies that the x-axis will represent the gender of the passengers, i.e., the values in the sex column (typically 'male' and 'female').

# hue='survived': This groups the data further within each gender by the survived column, which indicates whether the passenger survived (1) or not (0).

# The hue differentiates the bars by the survival status (i.e., a different color for those who survived and those who didnâ€™t).

# data=titanic: The dataset being used is the Titanic dataset.

# palette='pastel': This specifies the color palette for the plot. 'pastel' gives softer, lighter colors, making the chart visually appealing.

````

````python
sns.countplot(x= 'class', hue= 'survived', data= titanic, palette='pastel')
plt.xlabel('Class')
plt.ylabel('Number of People')
plt.legend(['Not Survived', 'Survived'])
plt.title('Survival Graph Based on Class')
plt.show()
````

````python
sns.histplot(data= titanic, x= 'age', hue= 'survived', bins = 20, kde=True, color='green', element='step')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.legend(['Not Survived', 'Not Survived','Survived', 'Survived'])
plt.title('Survival Graph Based on Age')
plt.show()

# bins refers to the number of intervals or "buckets" into which the data is divided when plotting a histogram
 # In a histogram, the x-axis is divided into intervals (bins), and the frequency of values that fall within each interval is displayed on the y-axis. The number of bins determines how granular or coarse the distribution appears.

# Definition: hue is used to group the data by a categorical variable and assign different colors to each group.

# Purpose: It allows you to distinguish data points by color, enabling you to see how a categorical variable influences the distribution of another variable. For example, in a survival analysis on Titanic passengers, the hue parameter can be used to differentiate between passengers who survived and those who didn't.
````

````python
sns.boxplot(data = titanic, x='survived', y= 'fare',hue= 'class', palette='pastel')
plt.xlabel('Survived')
plt.ylabel('Fare')
# plt.legend(['Not Survived', 'Survived'])
plt.title('Survival Graph Based on Fare')
plt.xticks([0,1], ['Not Survived', 'Survived'])
plt.show()

# Definition: palette defines the color scheme used to color the plots.

# Purpose: The palette allows you to choose a color set for different data groups. It can be a predefined color palette or a custom color palette. It is particularly useful when you want to ensure the visual clarity of the data or align the colors with a specific theme.
````

# Check how the price is distributed for each passenger by plotting a histogram.

````python
 # Plot a histogram to see how the price of the ticket ('fare') is distributed
plt.figure(figsize=(10,6))
sns.histplot(titanic['fare'], kde=True, bins=30)  # 'kde=True' adds a Kernel Density Estimate for smoother visualization
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

# Explanation: This line sets the figure size of the plot.

# figsize=(10, 6) specifies the width and height of the plot in inches (width=10 inches, height=6 inches). You can adjust these values to change the size of the plot to fit your needs.

# Explanation: This line creates the histogram using Seaborn's histplot function.

# titanic['fare']: This specifies the data to be plotted, specifically the 'fare' column from the Titanic dataset. It represents the ticket prices for each passenger.

# kde=True: This argument adds a Kernel Density Estimate (KDE) curve on top of the histogram. A KDE is a smoothed version of the histogram and helps visualize the distribution of the data more clearly. The KDE is drawn as a continuous line, giving a smooth estimate of the probability density function of the 'fare' values.

# bins=30: This defines the number of bins (intervals) into which the 'fare' values will be divided. Setting bins=30 means the range of the 'fare' values will be divided into 30 intervals (or buckets), and the height of each bar represents the frequency of values within each interval.
````

````python
# ### Explanation of Different Types of Graphs and Their Use

# 1. **Count Plot (`sns.countplot`)**:

#    * **When to Use**: Count plots are used to show the count of occurrences of categorical data. It gives you a quick view of how many instances of each category exist.

#    * **Syntax**: `sns.countplot(x='column_name', data=dataset, hue='another_column')`

#    * **Explanation**: A `countplot` is used to visualize the number of occurrences of a categorical variable. You can group these counts by another categorical variable using the `hue` parameter.

#    * **Example**: `sns.countplot(x='survived', data=titanic)` visualizes the survival status (0 or 1) of passengers on the Titanic. It helps us see how many passengers survived versus how many did not.

# 2. **Histogram Plot (`sns.histplot`)**:

#    * **When to Use**: Histograms are used to visualize the distribution of numerical data. You can see the frequency of different ranges of values.

#    * **Syntax**: `sns.histplot(data=dataset, x='column_name', bins=number_of_bins, kde=True)`

#    * **Explanation**: A histogram groups data into bins and shows the frequency (count) of data points within each bin. Adding `kde=True` superimposes a kernel density estimate (KDE), which is a smooth curve representing the distribution.

#    * **Example**: `sns.histplot(data=titanic, x='age', hue='survived', bins=20, kde=True)` shows how the ages of passengers are distributed, further breaking down the data by whether the person survived or not.

# 3. **Box Plot (`sns.boxplot`)**:

#    * **When to Use**: Box plots are used to display the distribution of numerical data through quartiles, highlighting the median, upper and lower quartiles, and outliers.

#    * **Syntax**: `sns.boxplot(x='category_column', y='numerical_column', data=dataset, hue='optional_group')`

#    * **Explanation**: A box plot is useful for understanding the spread and potential outliers of a numerical variable in a dataset. The box represents the interquartile range (IQR), the line in the box represents the median, and the "whiskers" represent the range of data excluding outliers.

#    * **Example**: `sns.boxplot(data=titanic, x='survived', y='fare', hue='class')` shows how the `fare` (ticket price) varies between passengers who survived and those who didn't, grouped by `class`. This gives insight into how different classes (First, Second, Third) might have affected survival chances.

# ### Explanation of the Specific Code Examples

# 1. **Survival Graph (`sns.countplot(x='survived', data=titanic)`)**:

#    * **Purpose**: This plot shows how many passengers survived and how many did not (binary survival status).
#    * **Explanation**: The x-axis represents the survival status (0 = Not Survived, 1 = Survived), and the y-axis represents the number of people in each category.
#    * **Customization**: `plt.xticks([0,1], ['Not Survived', 'Survived'])` changes the labels on the x-axis to make them more readable.

# 2. **Survival Graph Based on Gender (`sns.countplot(x='sex', hue='survived', data=titanic)`)**:

#    * **Purpose**: This plot shows the count of survivors and non-survivors for each gender (male, female).
#    * **Explanation**: The `hue='survived'` parameter colors the bars based on whether passengers survived or not, grouped by gender.
#    * **Customization**: `plt.legend(['Not Survived', 'Survived'])` customizes the legend to make it easier to understand.

# 3. **Survival Graph Based on Class (`sns.countplot(x='class', hue='survived', data=titanic)`)**:

#    * **Purpose**: This count plot helps us analyze how survival chances vary across different classes (First, Second, Third).
#    * **Explanation**: The `class` column categorizes passengers into different class groups, and survival status is differentiated by color using the `hue` parameter.
#    * **Note**: This plot can provide insight into whether passengers in first class had a higher survival rate compared to those in lower classes.

# 4. **Survival Graph Based on Age (`sns.histplot(data=titanic, x='age', hue='survived', bins=20, kde=True)`)**:

#    * **Purpose**: This histogram shows the distribution of ages of passengers, split by survival status.
#    * **Explanation**: The age of passengers is shown on the x-axis, and the number of people in each age group is plotted on the y-axis. The `hue` parameter splits the data by survival status, and the `kde=True` adds a smooth density curve.

# 5. **Survival Graph Based on Fare (`sns.boxplot(data=titanic, x='survived', y='fare', hue='class')`)**:

#    * **Purpose**: This box plot shows the distribution of fares for passengers who survived vs. those who didnâ€™t, further grouped by class.
#    * **Explanation**: The x-axis represents the survival status (0 = Not Survived, 1 = Survived), and the y-axis represents the fare of the passengers. The `hue='class'` separates the data based on passenger class, so you can see how the fare distribution differs within each class.

# ### Conclusion: When to Use Which Plot

# * **`countplot`**: Ideal for categorical data to show the frequency of occurrences. Use it when you need to compare the counts of categories.
# * **`histplot`**: Best for visualizing the distribution of numerical data. Use it when you want to understand the distribution of values (e.g., age, fare) within your dataset.
# * **`boxplot`**: Useful for displaying the distribution, spread, and outliers in numerical data. Use it when you want to compare distributions (e.g., fare) across categories (e.g., survival status, class).

# Each of these visualizations provides a unique perspective on the data, helping you identify patterns, trends, and outliers.
````

````python
# Hereâ€™s a comprehensive list of **major types of visualizations** you can use for different types of data exploration. Each plot type is categorized based on the type of data you are working with and the insights you aim to gain.

# ---

# ### **Categorical Data Visualization**

# 1. **Count Plot (`sns.countplot`)**

#    * **Purpose**: To visualize the count of observations in each categorical variable.
#    * **When to Use**: When you want to compare the frequency of categories in a dataset.
#    * **Example**: `sns.countplot(x='category', data=dataset)`

# 2. **Bar Plot (`sns.barplot`)**

#    * **Purpose**: To show the average (or other aggregate measures) of numerical data for each category.
#    * **When to Use**: When you want to compare the mean (or other statistical measures) of numerical variables across categories.
#    * **Example**: `sns.barplot(x='category', y='numerical_column', data=dataset)`

# 3. **Stacked Bar Plot**

#    * **Purpose**: To show proportions of categories within each category.
#    * **When to Use**: When comparing the parts that make up the whole within categorical variables.
#    * **Example**: `sns.barplot(x='category', y='count', hue='subcategory', data=dataset)`

# 4. **Pie Chart**

#    * **Purpose**: To visualize the proportion of categories.
#    * **When to Use**: When you want to display the percentage distribution of categorical variables.
#    * **Example**: `plt.pie(df['category'].value_counts())`

# ---

# ### **Numerical Data Visualization**

# 1. **Histogram Plot (`sns.histplot`)**

#    * **Purpose**: To show the distribution of a numerical variable.
#    * **When to Use**: To observe the frequency distribution of a continuous numerical variable.
#    * **Example**: `sns.histplot(data=dataset, x='numerical_column', bins=30)`

# 2. **Box Plot (`sns.boxplot`)**

#    * **Purpose**: To display the distribution of a numerical variable through quartiles.
#    * **When to Use**: When you want to visualize the spread and identify outliers in numerical data.
#    * **Example**: `sns.boxplot(x='category', y='numerical_column', data=dataset)`

# 3. **Violin Plot (`sns.violinplot`)**

#    * **Purpose**: To show the distribution and probability density of a numerical variable.
#    * **When to Use**: To compare distributions across categories, combining aspects of both box plots and kernel density estimates (KDE).
#    * **Example**: `sns.violinplot(x='category', y='numerical_column', data=dataset)`

# 4. **Density Plot (`sns.kdeplot`)**

#    * **Purpose**: To show the estimated probability density function of a numerical variable.
#    * **When to Use**: When you want to visualize the smooth distribution of a continuous variable.
#    * **Example**: `sns.kdeplot(data=dataset, x='numerical_column')`

# ---

# ### **Relationships Between Two Variables**

# 1. **Scatter Plot (`sns.scatterplot`)**

#    * **Purpose**: To show the relationship between two numerical variables.
#    * **When to Use**: To observe how two continuous variables relate to each other.
#    * **Example**: `sns.scatterplot(x='numerical_column_1', y='numerical_column_2', data=dataset)`

# 2. **Line Plot (`sns.lineplot`)**

#    * **Purpose**: To show trends in data over continuous intervals.
#    * **When to Use**: When you want to visualize continuous data points over time or any continuous sequence.
#    * **Example**: `sns.lineplot(x='time', y='value', data=dataset)`

# 3. **Pair Plot (`sns.pairplot`)**

#    * **Purpose**: To show pairwise relationships between multiple numerical variables.
#    * **When to Use**: When you want to explore the relationships between multiple numerical features in a dataset.
#    * **Example**: `sns.pairplot(dataset)` (plots scatter plots for each pair of columns)

# 4. **Joint Plot (`sns.jointplot`)**

#    * **Purpose**: To visualize the relationship between two numerical variables along with their marginal distributions.
#    * **When to Use**: To combine scatter plots and histograms, showing both the relationship and distribution of two variables.
#    * **Example**: `sns.jointplot(x='numerical_column_1', y='numerical_column_2', data=dataset)`

# 5. **Heatmap (`sns.heatmap`)**

#    * **Purpose**: To show matrix-style data, often used to represent correlations.
#    * **When to Use**: When you want to visualize the correlation between multiple variables or when dealing with a large matrix of data.
#    * **Example**: `sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')`

# ---

# ### **Categorical and Numerical Data Visualization**

# 1. **Facet Grid (`sns.FacetGrid`)**

#    * **Purpose**: To create a grid of plots based on the values of one or more categorical variables.
#    * **When to Use**: When you want to compare multiple subsets of data across categories.
#    * **Example**:

#      ```python
#      g = sns.FacetGrid(dataset, col='category_column')
#      g.map(sns.histplot, 'numerical_column')
#      ```

# 2. **Strip Plot (`sns.stripplot`)**

#    * **Purpose**: To show individual data points along with categorical variables.
#    * **When to Use**: When you want to see individual data points for a categorical variable.
#    * **Example**: `sns.stripplot(x='category', y='numerical_column', data=dataset)`

# 3. **Swarm Plot (`sns.swarmplot`)**

#    * **Purpose**: To show the distribution of data points in a category without overlapping.
#    * **When to Use**: When you want to visualize the distribution of individual data points in categorical data while avoiding overlap.
#    * **Example**: `sns.swarmplot(x='category', y='numerical_column', data=dataset)`

# 4. **Bar Plot with Error Bars**

#    * **Purpose**: To show categorical data along with the uncertainty (e.g., standard deviation, confidence intervals).
#    * **When to Use**: When you want to display the uncertainty around the mean value for each category.
#    * **Example**: `sns.barplot(x='category', y='numerical_column', data=dataset, ci='sd')`

# ---

# ### **Advanced Plots**

# 1. **Correlation Matrix**

#    * **Purpose**: To show the correlation between different numerical variables.
#    * **When to Use**: When you want to visually represent the correlation coefficients between variables in a matrix format.
#    * **Example**: `sns.heatmap(df.corr(), annot=True, cmap='coolwarm')`

# 2. **Radar Chart (Spider Plot)**

#    * **Purpose**: To visualize multivariate data across multiple axes.
#    * **When to Use**: When you want to compare multiple variables with a common origin.
#    * **Example**: Use libraries like `matplotlib` or `plotly` for radar charts. (Seaborn does not support radar charts directly.)

# 3. **3D Scatter Plot**

#    * **Purpose**: To visualize the relationship between three numerical variables.
#    * **When to Use**: When you need to visualize the relationship between three continuous variables.
#    * **Example**: `ax.scatter(x, y, z)` using `matplotlib`'s `Axes3D`.

# 4. **Area Plot**

#    * **Purpose**: To display quantities over time or another continuous variable, showing the cumulative total.
#    * **When to Use**: When you want to visualize the cumulative values for one or more variables over time or any continuous metric.
#    * **Example**: Use `plt.fill_between()` in `matplotlib`.

# ---

# ### **When to Use Which Plot?**

# 1. **Exploratory Data Analysis (EDA)**:

#    * **For understanding distributions**: Use **histogram**, **boxplot**, **violin plot**, and **density plot**.
#    * **For relationships between variables**: Use **scatter plot**, **line plot**, **pair plot**, **heatmap**, and **joint plot**.

# 2. **Categorical Data**:

#    * **For counting categories**: Use **count plot** and **bar plot**.
#    * **For comparing categories**: Use **bar plot**, **strip plot**, **swarm plot**, **violin plot**.

# 3. **Comparing Multiple Categories**:

#    * Use **FacetGrid** to create a grid of plots based on category values.

# ---

# This list covers the **major types of visualizations** and their use cases. Each type serves a different purpose based on whether you're working with categorical, numerical, or both types of data. Understanding when to use each plot will help you better explore your dataset and uncover important insights.
````
