```python
# 1. Import all the required Python Libraries

import pandas as pd
import numpy as np
import seaborn as sns #not needed here
import matplotlib.pyplot as plt #not needed here
```

```python
# 2. Locate an Open Source Data from the Web 

import os

# Check if iris.csv exists in the current directory
print("iris.csv found:", os.path.exists("iris.csv"))
print("Current directory:", os.getcwd())


# This dataset contains 150 records of iris flowers, with 5 attributes:

# SepalLengthCm

# SepalWidthCm

# PetalLengthCm

# PetalWidthCm

# Species (categorical label)

# (150,6) 

# Id is first column starting from 1
```

```python
#3. Load the Dataset into Pandas DataFrame

# url = "https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv"
# df = pd.read_csv(url)
df = pd.read_csv("iris.csv")
# read_csv : Read a comma-separated values (csv) file into DataFrame.
#File names should always be in quotes else it will take it as variable

# Show the first few rows (5) of the dataset to confirm successful loading
print(df.head())
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()


# Describe the dataset
dataset_description = df.describe()

# Display missing values and dataset description
print("Missing Values:\n", missing_values)
print("\nNot null values:\n", df.notnull().sum())
print("\nDataset Description:\n", dataset_description)

# Display the shape (dimensions) of the dataset
print("\nDataset Shape:", df.shape)



print("\nData Types of Variables:")
print(df.dtypes)
print(df.info()) 
```

```python
#5. Data Formatting and Data Normalization

# Check the data types of the columns
data_types = df.dtypes

print("\nData Types Before Conversion:")
print(df.dtypes)

# Convert specific float columns to integers
#first drop nan values before converting data types
df['SepalLengthCm'] = df['SepalLengthCm'].astype(int)
df['SepalWidthCm'] = df['SepalWidthCm'].astype(int)
df['PetalLengthCm'] = df['PetalLengthCm'].astype(int)
df['PetalWidthCm'] = df['PetalWidthCm'].round().astype(int)
# Note that during conversion, any decimal parts are discarded (for example, 5.8 becomes 5). If you need to round the values before converting to integer, you can use .round() before converting to int. 

# Display the updated DataFrame
print("\nDataFrame with Float Columns Converted to Integer:")
print(df.head())

# Check the data types of all columns after conversion
print("\nData Types After Conversion:")
print(df.dtypes)

print(df.info())
```

```python
# Display the first few rows of the dataframe
print("Original Data:")
print(df.head())

# Convert the 'Species' column to one-hot encoded columns
df_one_hot = pd.get_dummies(df, columns=['Species'], prefix=['Species'])


# Display the updated dataframe
print("\nData with One-Hot Encoding:")
print(df_one_hot.head())
```

```python
# Or can use : 

df["Species"].replace({"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2}, inplace=True)

# .replace(): The replace() method in pandas is used to replace specific values in a DataFrame or Series. Here, it takes a dictionary that defines which values should be replaced by what. The keys in the dictionary are the values that you want to replace, and the corresponding values are the replacements  

# inplace=True: This tells pandas to modify the original DataFrame directly rather than returning a modified copy. If inplace were set to False (or omitted, as False is the default), it would return a new DataFrame with the changes, leaving the original DataFrame unchanged 
```
